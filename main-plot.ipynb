```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker
from mpl_toolkits.axes_grid.inset_locator import inset_axes
from readData import *

import sys
sys.path.append("../library/")
import defaultPlot
from stringFormat import *
```

```python
networkSize = 1000
meanDegree = 10
SI_II = 0.2
I_R = 0.2
randomEngineSeed = 0
```

```python
_, ax = plt.subplots(figsize=(10,10))


networkType, spreadType = "RR", "RK4"
time, S, I, R = readSIR(networkType, spreadType, networkSize, meanDegree, SI_II, I_R, randomEngineSeed)
ax.plot(time, S, color='g', linestyle='-', label=latexString("S(RR-RK4)"))
ax.plot(time, I, color='r', linestyle='-', label=latexString("I(RR-RK4)"))
ax.plot(time, R, color='b', linestyle='-', label=latexString("R(RR-RK4)"))


networkType, spreadType = "ER", "GA_sync"
time, S, I, R = readSIR(networkType, spreadType, networkSize, meanDegree, SI_II, I_R, randomEngineSeed)
ax.plot(time, S, color='g', linestyle='--', label=latexString("S(ER-GA-sync)"))
ax.plot(time, I, color='r', linestyle='--', label=latexString("I(ER-GA-sync)"))
ax.plot(time, R, color='b', linestyle='--', label=latexString("R(ER-GA-sync)"))

networkType, spreadType = "ER", "GA_async"
time, S, I, R = readSIR(networkType, spreadType, networkSize, meanDegree, SI_II, I_R, randomEngineSeed)
ax.plot(time, S, color='g', linestyle='-.', label=latexString("S(ER-GA-async)"))
ax.plot(time, I, color='r', linestyle='-.', label=latexString("I(ER-GA-async)"))
ax.plot(time, R, color='b', linestyle='-.', label=latexString("R(ER-GA-async)"))


networkType, spreadType = "ER", "DT"
time, S, I, R = readSIR(networkType, spreadType, networkSize, meanDegree, SI_II, I_R, randomEngineSeed)
ax.plot(time, S, color='g', linestyle=':', label=latexString("S(ER-DT)"))
ax.plot(time, I, color='r', linestyle=':', label=latexString("I(ER-DT)"))
ax.plot(time, R, color='b', linestyle=':', label=latexString("R(ER-DT)"))


ax.legend(loc="upper right", fontsize=20)
ax.set_xlabel(latexString("t"))
ax.set_ylabel(latexString("\\rho"))
ax.set_xlim(left=0)
ax.set_ylim([0,1])


plt.show()
```
